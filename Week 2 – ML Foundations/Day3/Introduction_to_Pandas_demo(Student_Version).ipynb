> ### Read in the telecom churn dataset.

> ### 1. Check for how many numerical and categorical there are in the dataset.
>    - Use the select_dtypes() method. [select_dtypes](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.select_dtypes.html)

```python
import pandas as pd

df = pd.read_csv('telecom.csv')
print("Dataset loaded ✓")
```

```python
df.head()
```

```python
print("Shape:", df.shape)
print(f"Rows: {df.shape[0]:,}  |  Columns: {df.shape[1]}")

# Null values
print("\n--- Null Values ---")
print(df.isnull().sum())
print(f"\nTotal nulls: {df.isnull().sum().sum()}")
```

```python
# Numerical columns
numerical = df.select_dtypes(include=['number'])
print(f"Numerical columns ({len(numerical.columns)}):")
print(numerical.columns.tolist())

# Categorical columns
categorical = df.select_dtypes(include=['object', 'category'])
print(f"\nCategorical columns ({len(categorical.columns)}):")
print(categorical.columns.tolist())

print(f"\nSummary: {len(numerical.columns)} numerical, {len(categorical.columns)} categorical")
```

> ### 2. Map the following columns `Partner`, `Dependents`, `PhoneService`
>    - For each of the columns listed above, map `Yes` to `1`, `No` to `0`.


```python
# for loop
# map
#replace
```

```python
# Columns to map
cols = ['Partner', 'Dependents', 'PhoneService']

# Map Yes → 1, No → 0
df[cols] = df[cols].replace({'Yes': 1, 'No': 0})

# Verify the result
print(df[cols].head(5))
print("\nUnique values after mapping:")
for col in cols:
    print(f"  {col}: {sorted(df[col].unique())}")
```

> ### 3. Convert the data type of the `TotalCharges` column to the float data type.
>    - Replace missing values with 0

```python
# Step 1: Convert to numeric (coerce turns invalid strings into NaN)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Step 2: Replace NaN with 0
df['TotalCharges'] = df['TotalCharges'].fillna(0)

# Verify
print("dtype :", df['TotalCharges'].dtype)
print("Nulls :", df['TotalCharges'].isnull().sum())
print(df['TotalCharges'].head(5))
```

> ### 4. Return the top 10 customers with the highest `TotalCharges`.
>    - a) How many of these customers have multiple lines?
>    - b) How many of these customers have subscribed to `PaperlessBilling`?

```python
# Get top 10 customers by TotalCharges
top10 = df.nlargest(10, 'TotalCharges')

print(top10[['customerID', 'TotalCharges', 'MultipleLines', 'PaperlessBilling']])
```

```python
# a) How many have multiple lines?
multiple_lines = top10[top10['MultipleLines'] == 'Yes']
print(f"Customers with Multiple Lines: {len(multiple_lines)}")
print(multiple_lines[['customerID', 'TotalCharges', 'MultipleLines']])
```

```python
# b) How many subscribed to PaperlessBilling?
paperless = top10[top10['PaperlessBilling'] == 'Yes']
print(f"Customers with PaperlessBilling: {len(paperless)}")
print(paperless[['customerID', 'TotalCharges', 'PaperlessBilling']])
```

> ### 5. Return the top 100 customers with the highest `MonthlyCharges`
>    - a) What was the most frequent `InternetService` used by these customers?
>    - b) How many of these customers churned? (Return the results as a percentage)
>        - **hint:** `normalize` parameter in value_counts() method.

```python
# Get top 100 customers by MonthlyCharges
top100 = df.nlargest(100, 'MonthlyCharges')

print(top100[['customerID', 'MonthlyCharges', 'InternetService', 'Churn']].head(10))
```

> ### 6. Create a new column named `loyalty`. This column will contain the following:
> Iterate through the `tenure` column and store the following elements in a list. Assign the list to the column `loyalty`:
>    - If the customers tenure is below 12, flag it as `Bronze`.
>    - If the customers tenure is between 12 and 24, flag it as `Silver`.
>    - If the customers tenure is above 24, flag it as `Gold`.


```python
df['tenure']
```

```python
loyalty = []

for t in df['tenure']:
    if t < 12:
        loyalty.append('Bronze')
    elif t <= 24:
        loyalty.append('Silver')
    else:
        loyalty.append('Gold')

df['loyalty'] = loyalty
```

```python
df[['tenure', 'loyalty']].head()
```

> ### 7. Create a new column named `Customer Value`. This column will contain the following:
> Iterate through the `TotalCharges` column and store the following elements in a list. Assign the list to the column `Customer Value`:
>    - If the customers TotalCharges is below 1000, flag the customer as `low`.
>    - If the customers TotalCharges is between 1000 and 3000, flag the customer as `medium`.
>    - If the customers TotalCharges is above 3000, flag the customer as `high`.

```python

# Iterate through TotalCharges and build the list
customer_value = []

for charge in df['TotalCharges']:
    if charge < 1000:
        customer_value.append('low')
    elif 1000 <= charge <= 3000:
        customer_value.append('medium')
    else:
        customer_value.append('high')

# Assign the list to the new column
df['Customer Value'] = customer_value

# Verify
print(df[['TotalCharges', 'Customer Value']].head(10))
print("\nValue counts:")
print(df['Customer Value'].value_counts())
```

> ### 8. Which payment method is the most frequent?
>    - a) Use the to_dict() method to convert the output to a dictionary.
>    - b) Create a new dataframe `top_payment_method` using the dictionary generated from the previous step. Name the column `frequency`.

```python
# Get value counts and find the most frequent
payment_counts = df['PaymentMethod'].value_counts()
print("Payment Method Counts:\n", payment_counts)
print(f"\nMost frequent: {payment_counts.idxmax()} ({payment_counts.max()} customers)")
```

```python
# a) Convert value_counts output to a dictionary
payment_dict = payment_counts.to_dict()
print(payment_dict)
```

```python
# b) Create a new dataframe from the dictionary
top_payment_method = pd.DataFrame.from_dict(
    payment_dict,
    orient='index',
    columns=['frequency']
)

print(top_payment_method)
```

> ### 9. How many customers are on a One year contract and have not completed a 12 month tenure?

```python
# Two conditions combined with &
mask = (df['Contract'] == 'One year') & (df['tenure'] < 12)

result = df[mask]

print(f"Count: {len(result)}")
print(result[['customerID', 'Contract', 'tenure']])
```

## Exploring customers who churned.
>    - Create a dataframe for customers who churned and name the dataframe `churned`

```python
# Filter only customers who churned
churned = df[df['Churn'] == 'Yes']

# Verify
print(f"Shape: {churned.shape}")
print(f"Total churned customers: {len(churned)}")
churned.head()
```

> ### 12. For customers who churned:
>    - a) What is the gender distribution?
>    - b) What is the average tenure length per `Contract`?
>    - c) For Males & Females, what is the average tenure length per `Contract` for each gender group?
>    - d) For Males & Females, what is the average tenure length per `Contract` for each gender group for customers who are on `DSL`?

```python
# a) Gender distribution of churned customers
gender_dist = churned['gender'].value_counts()
print("Gender Distribution:\n", gender_dist)
print("\nAs percentage:")
print((churned['gender'].value_counts(normalize=True) * 100).round(2))
```

```python
# b) Average tenure per Contract type
avg_tenure_contract = churned.groupby('Contract')['tenure'].mean().round(2)
print("Avg Tenure per Contract:\n", avg_tenure_contract)
```

```python
# c) Average tenure per Contract grouped by gender
avg_tenure_gender = churned.groupby(['gender', 'Contract'])['tenure'].mean().round(2)
print("Avg Tenure by Gender & Contract:\n", avg_tenure_gender)
```

```python
# d) Filter DSL customers first, then group by gender & Contract
dsl_churned = churned[churned['InternetService'] == 'DSL']

avg_tenure_dsl = dsl_churned.groupby(['gender', 'Contract'])['tenure'].mean().round(2)
print("Avg Tenure by Gender & Contract (DSL only):\n", avg_tenure_dsl)
```
